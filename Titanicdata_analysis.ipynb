```python

# 1. Convert categorical to numeric
# Fill missing Embarked with mode
titanic_data['Embarked'] = titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0])

# Encode 'Sex' and 'Embarked' using LabelEncoder
le = LabelEncoder()
titanic_data['Sex'] = le.fit_transform(titanic_data['Sex'])  # male=1, female=0
titanic_data['Embarked'] = le.fit_transform(titanic_data['Embarked'])  # S=2, C=0, Q=1

# 2. Handle missing values
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())
titanic_data['Fare'] = titanic_data['Fare'].fillna(titanic_data['Fare'].median())
titanic_data['Cabin'] = titanic_data['Cabin'].fillna('Missing')

# 3. Create new features
# FamilySize = SibSp + Parch + 1
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1

# IsAlone = 1 if FamilySize == 1, else 0
titanic_data['IsAlone'] = (titanic_data['FamilySize'] == 1).astype(int)

# Extract Title from Name
titanic_data['Title'] = titanic_data['Name'].str.extract(' ([A-Za-z]+)\\.', expand=False)

# Simplify titles
titanic_data['Title'] = titanic_data['Title'].replace(
    ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 
     'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
titanic_data['Title'] = titanic_data['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs'})

# Encode Title
titanic_data['Title'] = le.fit_transform(titanic_data['Title'])

# 4. Scale numerical features, but **don't scale 'FamilySize'**
scaler = StandardScaler()
numeric_features = ['Fare']  # Only scale 'Fare'

# Apply scaling
titanic_data[numeric_features] = scaler.fit_transform(titanic_data[numeric_features])

# 5. Remove Negative Values: Shift scaled columns to non-negative values
for feature in numeric_features:
    min_value = titanic_data[feature].min()
    if min_value < 0:
        titanic_data[feature] = titanic_data[feature] - min_value  # Shift values to be non-negative

# (Optional) Drop irrelevant columns for modeling
titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin','AgeGroup'], axis=1, inplace=True)

# Preview the processed data
print(titanic_data.head())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Features and target
X = titanic_data.drop('Survived', axis=1)
y = titanic_data['Survived']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)

print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, lr_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, lr_preds))
print("Classification Report:\n", classification_report(y_test, lr_preds))
```

```python

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)

print("Random Forest:")
print("Accuracy:", accuracy_score(y_test, rf_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_preds))
print("Classification Report:\n", classification_report(y_test, rf_preds))
```
